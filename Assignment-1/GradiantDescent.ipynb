**TEST CASE 1**

For: function: f(x) = x**2+5  

learning rate 0.1

```
import sympy as sym
from sympy import *
import random
#step1
#initialize
X = random.randint(1,10) #generates random value for X between 1 and 10
print("Random X value generated: ",X)
epochs=20
iter=1
n=0.1
x=0
x=sym.Symbol('x')
f = x**2+5
while 1:
  print("Iteration number : {}".format(iter))
  #step2
  #Calculate Gradient/slope/first order derivative of objective function
  #f = x**4+3*x**2+10
  #import sympy as sym
  #declare a variable that you will use as a letter in your function
  x=sym.Symbol('x')
  df=sym.diff(f)
  slope = df.subs(x,X)
  print("Step2: slope {}".format(slope))
  #Step3 : Find the change in variable value
  delta_X = -(n)*slope
  print("Step3: delta_X {}".format(delta_X))
  #Step4 : Updating variable value
  X = X+delta_X
  print("Step4: Updated value X = {}".format(X))
  Y = f.subs(x,X)
  print("Checking f(X) value when substituted updated X value f(updated_X) = {}".format(Y))
  #Step5 : update iteration
  print("Step 5 - Update iteration")
  iter = iter+1
  if(iter>epochs):
    break
  else:
    print("Goto step 2")
    print("-------------------------------------------")
print("FINAL RESULT_-------------------------------------------------------------------")
print("Minimum value of point X is",X)
print("Minimum value is ",f.subs(x,X))
print('Lets round/floor the values obtained:')
Y= f.subs(x,X);
print("---- Minima point:----")
print("X: ",round(X,2))
print("Y: ",round(Y,2))
```
